# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
rides_city_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
rides_city_type.head()
```

```python
# 2. Get the total drivers for each city type
drivers_city_type = city_data_df.groupby(["type"]).sum()["driver_count"]
drivers_city_type.head()
```

```python
#  3. Get the total amount of fares for each city type
fares_city_type = pyber_data_df.groupby(["type"]).sum()["fare"]
fares_city_type.head()
```

```python
#  4. Get the average fare per ride for each city type.
fare_ride_avg = fares_city_type/rides_city_type
fare_ride_avg
```

```python
# 5. Get the average fare per driver for each city type. 
fare_driver_avg = fares_city_type/drivers_city_type
fare_driver_avg
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({
        "Total Rides" : rides_city_type,
        "Total Drivers" : drivers_city_type,
        "Total Fares" : fares_city_type,
        "Average Fare Per Ride" : fare_ride_avg,
        "Average Fare Per Driver" : fare_driver_avg})

pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:.0f}".format)

pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:.0f}".format)

pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:.2f}".format)

pyber_summary_df["Average Fare Per Ride"] = pyber_summary_df["Average Fare Per Ride"].map("${:.2f}".format)

pyber_summary_df["Average Fare Per Driver"] = pyber_summary_df["Average Fare Per Driver"].map("${:.2f}".format)

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 9. Using groupby() on the pyber_data_df to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
city_type_dataframe = pd.DataFrame(pyber_data_df.groupby(["date","type"]).sum()["fare"])
city_type_dataframe
```

```python
# 10. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
city_type_dataframe = city_type_dataframe.reset_index()
```

```python
# 11. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
city_type_dataframe.pivot_table(values='fare', index='date', columns='type')
```

```python
# 12. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2018-01-01':'2018-04-29'.
early2019_df = pd.DataFrame(city_type_dataframe.loc[(city_type_dataframe["date"] > "1/1/2019") & (city_type_dataframe["date"] <= "4/29/2019")])
early2019_df = early2019_df.pivot( values = "fare", index = "date", columns = "type")
```

```python
# 13. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
early2019_df.index = pd.to_datetime(early2019_df.index)
early2019_df
```

```python
# 14. Check that the datatype for the index is datetime using df.info()
early2019_df.info()
```

```python
# 15. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_fares = early2019_df.resample("W").sum()
weekly_fares
```

```python
# 16. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

# Using the Object-Oriented approach to plot the weekly_fares DataFrame from step 15.
ax = weekly_fares.plot(figsize = (15,10))
# Setting Table Properties 
ax.set_title("Total Fare by City Type", fontsize = "20")
ax.set_xlabel("Year & Month", fontsize = "16")
ax.set_ylabel("Fares ($USD)", fontsize = "16")
ax.grid(True)
ax.legend(["Rural", "Suburban", "Urban"], title = "type", loc = "center", fontsize = "15", mode = "Expanded")
plt.savefig("analysis/PyBer_fare_summary.png")
plt.show()
```
